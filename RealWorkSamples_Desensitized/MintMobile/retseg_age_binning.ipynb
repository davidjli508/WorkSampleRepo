```python
def plotForBlockGroup(df_none, df_prior, df_after, blockgrouplist):
    for blkgrp in blockgrouplist:
        df_none_small = df_none[df_none["BLOCK_GROUP"] == blkgrp]
        df_prior_small = df_prior[df_prior["BLOCK_GROUP"] == blkgrp]
        df_after_small = df_after[df_after["BLOCK_GROUP"] == blkgrp]

        for df in [df_none_small, df_prior_small, df_after_small]:
            block_group_name = df.iloc[0]['BLOCK_GROUP']
            df = df.drop(columns = ["BLOCK_GROUP", "TOTAL_POP"])
            df = df.iloc[:, ::-1]

            values = df.iloc[0, 1:].values
            labels = df.columns[1:]

            # Plot the data
            plt.figure(figsize=(8, 5))
            plt.plot(labels, values, marker='o', linestyle='-')
            plt.xlabel('Age Group')
            plt.ylabel('Value')
            plt.title("Block Group: " + str(block_group_name))
            plt.xticks(rotation=90)  # Rotate x-axis labels by 45 degrees
            plt.show()
```
